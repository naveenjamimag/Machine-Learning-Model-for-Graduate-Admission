<b> Cleaning Column data </b>
- Removing Null values
- Removing noise
- Conversion of % and 10 pinter score in CGPA to 4 pointer
- Toefl and IELTS score to the same scale according to the information available on ETS Official website (https://www.ets.org/toefl/institutions/scores/compare/)

```python
def replace_special_chars(i):
    #a = re.sub('[^A-Za-z]+',' ',str(i))
    a=re.findall(r'\d+', str(i))
    #a = a.lower()
    return ''.join(a)
```

```python
df_all_neu['gre_score']=df_all_neu.gre_score.apply(replace_special_chars)
```

```python
df_all_neu['gre_score_quant']=df_all_neu['gre_score_quant'].apply(replace_special_chars)
```

```python
df_all_neu['test_score_toefl'] = df_all_neu['test_score_toefl'].apply(replace_special_chars)
df_all_neu['gre_score_verbal'] = df_all_neu['gre_score_verbal'].apply(replace_special_chars)
df_all_neu['work_ex'] = df_all_neu['work_ex'].apply(replace_special_chars)
```

```python
df_all_neu["undergraduation_score"] = [x.replace('CGPA','') for x in df_all_neu["undergraduation_score"]]
df_all_neu["undergraduation_score"] = [x.replace('%','') for x in df_all_neu["undergraduation_score"]]
df_all_neu["papers_published"] = [str(x).replace('Tech Papers','') for x in df_all_neu["papers_published"]]
```

```python
df_all_neu.dtypes
```

```python
df_all_neu.loc[df_all_neu['work_ex']=='','work_ex']=0
```

```python
df_all_neu=df_all_neu[df_all_neu.undergraduation_score!='Dream Univ.Virg. Tech']
```

```python
values=[]
for each in df_all_neu.undergraduation_score.unique():
    try:
        float(each)
    except:
        values.append(each)
```

```python
for each in values:
    df_all_neu=df_all_neu[df_all_neu.undergraduation_score!=each]
```

```python
df_all_neu[['gre_score','gre_score_quant','gre_score_verbal','test_score_toefl','undergraduation_score','work_ex']]=df_all_neu[['gre_score','gre_score_quant','gre_score_verbal','test_score_toefl','undergraduation_score','work_ex']].apply(pd.to_numeric)
```

```python
df_all_neu=df_all_neu.loc[~(df_all_neu.test_score_toefl.isna()),:]
```

```python
df_all_neu.isna().sum()
```

```python
df_all_neu.reset_index(drop=True,inplace=True)
```

```python
update_cgpa_score_scale_4 = []
for score in df_all_neu.undergraduation_score.tolist():
   s = 0
   try:
       score = float(score)
   except:
       score= 0
   if score > 10:
       s = ((score)/20) - 1
       s = round(s,2)
       update_cgpa_score_scale_4.append(s)
   else:
       s = ((score)/10)*4
       s = round(s,2)
       update_cgpa_score_scale_4.append(s)
```

```python
df_all_neu['undergraduation_score']=update_cgpa_score_scale_4
```

```python
 df_all_neu.loc[df_all_neu['test_score_toefl']<9,'test_score_toefl']=pd.cut(df_all_neu.loc[df_all_neu['test_score_toefl']<9,'test_score_toefl'], bins=[-1,0.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9], labels=[0,31,34,45,59,78,93,101,109,114,117,120])
```
