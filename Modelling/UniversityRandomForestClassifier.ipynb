```python
# Required Python Packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import pickle
```

     Importing the dataset

```python
dataset = pd.read_excel('./DataSet/FinalDataSetNortheastern.xlsx')
```

```python
dataset.head()
```

For predicting, if a student will get admitted to northeastern university, the columns which would be used are as follows
Labels: gre_score,gre_score_quant,gre_score_verbal,term_applying,test_score_toefl,undergraduation_score,work_ex,year_applying
Target: Status
As the term_applying is categorical, one-hot encoding will be done for it.

```python
#One-hot encoding 
def encode_and_bind(dataset, term_applying):
    term_hotencoded = pd.get_dummies(dataset[['term_applying']])
    res = pd.concat([dataset,term_hotencoded], axis=1)
#     res = res.drop(dataset['term_applying'], axis=1)
    return(res) 
```

```python
dataset_encoded=encode_and_bind(dataset,dataset['term_applying'])
```

```python
dataset_encoded.head()
```

Defining labels-X and Tragets-Y

```python

X=dataset_encoded[['gre_score','gre_score_quant','gre_score_verbal','test_score_toefl','undergraduation_score','work_ex','year_applying','term_applying_Fall','term_applying_Spring']].copy()
Y=dataset_encoded[['status']].copy()
```

```python
dataset_encoded.head()
```

Splitting the dataset_encoded into training and testing dataset 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

Predicting by using RandomForestClassifier

```python
def modeltraining(X_train,X_test,Y_train,Y_test):
    model=RandomForestClassifier()
    model.fit(X_train,Y_train)
    predicted_labels=model.predict(X_test)
    accuracy=accuracy_score(Y_test,predicted_labels)
    return accuracy,model
```

Accuracy for RandomForest Classifier

```python
accuracy,model=modeltraining(X_train,X_test,Y_train,Y_test)
```

Pickling the model

```python
# save the model to disk

filename = 'UniversityRandomForestPickled.sav'
pickle.dump(model, open(filename, 'wb'))
```
